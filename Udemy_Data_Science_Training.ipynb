```python
# Object oriented programming perks:
# 1) Modularity: Object-oriented programming provides object-based separation of code and behavior
# 2) Extensibility: Objects can be extended to include new behaviours
# 3) Reusability: objects can be reused everywhere in the application

# this is the class - so the blueprint
# Encapsulation - it is a fundamental feature of OOP
# we are able to group certain variables and functions
class Person:
  name = 'Kevin'
  # so this is the so-called constructor (initialization block)
  def __init__(self, name, age = 20):
    self.name = name
    self.age = age

  def show_name(self):
    print('My name is %s and age is %s ' % (self.name, self.age))

# we can instantiate a class: and p is the object
p = Person(name = 'Mahmut', age = 36)
p.show_name()
```

```python
# Class variables: shared by all object instances of the given class - the class own the variable
# Instance variable: the value of these variables can be different depending on the instance
class Person:
  gender = 'Male'  # This is a class variable which is the same for every instantiation
  person_list = [] # It is a dynamic class variable
  # so this is the so-called constructor (initialization block)
  def __init__(self, name, age = 20):
    self.name = name
    self.age = age

  def show_name(self):
    print('My name is %s and age is %s and gender is %s ' % (self.name, self.age, self.gender))

  def add_value(self, value):
    self.person_list.append(value)

p1 = Person(name = 'Kevin')
p2 = Person(name = 'Joe', age = 15)

p1.add_value('person1')
p1.add_value('person2')

p1.show_name()
p2.show_name()

print(p1.person_list)
print(p2.person_list)
```

```python
# Inheritance allows us to define a class (child class or derived class) that inherits all the properties and functions from another class
class Algorithm:

  # class variable
  general_message = 'This has something to do with algorithms...'

class SortingAlgorithm(Algorithm):

  # class variable
  sorting_message = 'This is a sorting algorithm...'

sorting_algorithm = SortingAlgorithm()
print(sorting_algorithm.general_message)
```

```python
# Super keyword
class User:

  # so this is the so-called constructor (initialization block)
  def __init__(self, name, age = 20):
    self.name = name
    self.age = age

  def show_user(self):
    print("The user is: %s and the age is: %s" % (self.name, self.age))

  def show_info(self):   # This is from the parent class
    print("This is from the parent...")

class FacebookUser(User):

  # super keywork is going to call the parent class
  def __init__(self, name, age):
      super().__init__(name, age)

  def show_info(self):   # This is from the child class (Function Override: it is present in the parent class)
    print("This is from the child...")

facebook_user = FacebookUser('Adam Smith', 35)
facebook_user.show_user()
facebook_user.show_info()
```

```python
# Polymorphism is the ability of an object to take on many forms
# In object-oriented programming, it can be achieved by inheritance
class SortingAlgorithm: # Parent Class

  def sort(self):
    pass


class InsertionSort(SortingAlgorithm):

  # function override
  def sort(self):
    print('Insertion sort is running...')

class SelectionSort(SortingAlgorithm):

  # function override
  def sort(self):
    print('Selection sort is running...')

class QuickSort(SortingAlgorithm):

  # function override
  def sort(self):
    print('Quick sort is running...')

# We ask the user what sorting algortihm to use
num = input("Please, select a sortin algorithm")
num = int(num)

# this is how we create a NULL object
sorting_algorithm = None

if num == 0:
  sorting_algorithm = InsertionSort()
elif num == 1:
  sorting_algorithm = SelectionSort()
elif num == 2:
  sorting_algorithm = QuickSort()
else:
  num = input("Please enter valid number")
  num = int(num)

# this object can take on many forms
sorting_algorithm.sort()
```

```python
# Vehicle Parent Class and Bus and Car Child Class
# Super keyword
class Vehicle:

  # so this is the so-called constructor (initialization block)
  def __init__(self, brand, year = 2020):
    self.brand = brand
    self.year = year

  def show_vehicle(self):
    print("Vehicle is: %s and the year is: %s" % (self.brand, self.year))

  def fuel_consumption(self):
    print("Vehicle is: %s and the year is: %s" % (self.brand, self.year))

class Bus(Vehicle):

  # super keywork is going to call the parent class
  def __init__(self, brand, year):
      super().__init__(brand, year)

  def fuel_consumption(self):
    print("This bus burns 2 gallons per mile")

class Car(Vehicle):

  # super keywork is going to call the parent class
  def __init__(self, brand, year):
      super().__init__(brand, year)

  def fuel_consumption(self):
    print("This bus burns 1 gallon per mile")

car1 = Car(brand = 'Ford', year = 2021)
car1.show_vehicle()
car1.fuel_consumption()
```

```python
# Comparing objects
class Person:

  def __init__(self, name, age):
    self.name = name
    self.age = age
  def __eq__(self, other): # We override equal function so that we can define the equivalance
    return self.age == other.age

  def __lt__(self, other): # We override equal function so that we can define less than
    return self.age < other.age

# how to compare the objects?
p1 = Person('Adam', 32)
p2 = Person('Joe', 34)

print(p1 == p2)
print(p1 < p2)
```

# **2.2.1 Types of Machine Learning Problems**

1.   **Problem Definition:** Aligning the problem you trying to solve with the machine learning problem.
2.   **Data**
  *   Structured Data: Data with rows and columns such as Excel or CSV data
  *  Unstructured Data: Free writing such as e-mail, Picture, Audio
3.   **Evaluation:** What defines success for us? we set the accuracy of our model depending on the problem. Typical metrics to be used:
  *   **Classification:** Accuracy, Precision, Recall
  *   **Regression:** Mean absolute error (MAE), Mean squared error (MSE), Root mean squared error (RMSE)
  *   **Recommmendation:** Precision at K
4.   **Features:** also referred to as feature variables
  *  Numerical features
  *  Categorical features:  Yes or No, Color, Model
5.   **Modelling:** You divide your data into *Training Data* where you train your model, *Validation Data* where you tune your model on this , and *Test Data* where you commpare on this
  *   Choosing and training a model
  *   Tuning a model
  *   Model comparison
6.   **Experimatation:** Simply try a new model to see if you get more accurate model




**Overfitting and Underfitting:**

*   **Overfittig:** Great performance on the training data but poor performance on test data means your model doesn't generalize well. Your model may be overfitting the training data. Try using a simpler model or making sure your the test data is of the same style your model is training on.Another form of overfitting can come in the form of better performance on test data than training data. This may mean your testing data is leaking into your training data (incorrect data splits) or you've spent too much time optimizing your model for the test set data. Ensure your training and test datasets are kept separate at all times and avoid optimizing a models performance on the test set
*   **Underfitting:** Poor performance on training data means the model hasn't learned properly and is underfitting. Try a different model, improve the existing one through hyperparameter or collect more data

## 2.6.3 Pick a model/estimator (to suit your problem)

Some things to note:

* Sklearn refers to machine learning models, algorithms as estimators
* Classification problem - perdicting a category
  * Sometimes you'll see `clf` (short for classifier) used as a classification estimator
* Regression problem - predicting a number

While picking and comparing models, we compare them based on their accuracy one of the most commonly used metris is $R^2$ value:
\begin{equation}
R^2 =  1 - \frac{\text{Unexplained Variation}}{\text{Total Variation}} = 1-\frac{\sum(y_i - \hat{y}_i)^2}{\sum(y_i - \bar{y}_i)^2}
\end{equation}
where $y_i$ is the actual value of target variable, $\hat{y}_i$ is the estimated value of target variable, and $\bar{y}_i$ is the mean value of target variable. Here, we try to see how much difference our makes rather than taking the mean of data. However, the model comparison metrics are not limited to $R^2$ value, other metrics include ***Confustion Matrix*** and ***ROC Curve***.


**Little guidance:**

1.   If you have structured data, use ensemble methods
2.  If you have unstructured data, use deep learning or transfer learning
